# 4. / 5. Transformacja danych i analiza relacji między zmiennymi

```python
"""
Biblioteka scikit-learn (często importowana jako sklearn) jest fundamentalną biblioteką w ekosystemie Pythona do uczenia maszynowego (Machine Learning). 
Główne Funkcje i Zastosowania

scikit-learn to potężne narzędzie, które dostarcza prostych i efektywnych narzędzi do:

    Klasyfikacji (Classification): Przewidywanie, do której kategorii należy dany element (np. czy mail jest spamem, czy zdjęcie przedstawia psa czy kota).

        Przykładowe algorytmy: Lasy losowe (Random Forests), SVM (Support Vector Machines), Regresja logistyczna.

    Regresji (Regression): Przewidywanie wartości ciągłej (np. ceny domu, temperatury, prognozy sprzedaży).

        Przykładowe algorytmy: Regresja liniowa (Linear Regression), Regresja grzbietowa (Ridge), Lasso.

    Klastrowania (Clustering): Grupowaanie podobnych elementów bez wstępnych etykiet (np. segmentacja klientów na podstawie ich zachowań).

        Przykładowe algorytmy: K-Means, DBSCAN.

    Selekcji Modeli (Model Selection): Narzędzia do optymalizacji i walidacji modeli (np. train_test_split do podziału danych, walidacja krzyżowa, wyszukiwanie najlepszych hiperparametrów za pomocą GridSearchCV).

    Preprocessingu Danych (Preprocessing): Transformacja danych, aby były lepiej przygotowane do uczenia maszynowego (np. skalowanie danych za pomocą StandardScaler, kodowanie zmiennych kategorycznych za pomocą OneHotEncoder).

Dlaczego Jest Kluczowa dla Analityka Danych?

    Spójny Interfejs (API): Wszystkie algorytmy w scikit-learn mają ten sam interfejs (metody fit(), predict(), transform()). To sprawia, że zmiana algorytmu na inny jest niezwykle prosta i szybka.

    Integracja: Bezproblemowo współpracuje z Pandas (do wczytywania i manipulacji danymi) oraz NumPy (do obliczeń numerycznych i przechowywania danych).

    Wydajność: W dużej mierze jest zbudowana na NumPY i SciPy, co zapewnia wysoką wydajność obliczeniową.

Podsumowanie Użycia

Gdy pracujesz z biblioteką Pandas nad oczyszczonymi danymi, scikit-learn to kolejny krok, który pozwala Ci:

    Wczytać dane do modelu.

    Przetworzyć je (skalowanie, normalizacja).

    Wytrenować model.

    Ocenić jego wydajność.

    Wykorzystać go do przewidywania nowych danych. 
    * z pomocą gemini
"""
# A więc z niej skorzystam

from sklearn.datasets import load_iris
# Pierwsza linijka kodu importuje funkcję load_iris() z biblioteki sklearn.datasets. Ta funkcja jest używana do wczytania zestawu danych zawierającego informacje o irysach.

iris = load_iris()
# Ta linijka kodu wywołuje funkcję load_iris() i przechowuje zwrócone dane w zmiennej iris.
# Zwrócone dane są typem Bunch, co oznacza, że są one podobne do słowników, ale można do nich dostępować za pomocą atrybutów.

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
# Tutaj tworzymy obiekt DataFrame biblioteki pandas, który jest dwuwymiarowym zestawem danych. 
# W założeniu jest to jedna z najważniejszych struktur tej biblioteki. 
# Potem zmienna df przechowuje ten obiekt. Do DataFrame przekazujemy iris.data jako dane, a iris.feature_names jako nazwy kolumn.

df['class'] = iris.target
# Dodaje nową kolumnę o nazwie 'class' do DataFrame df i wypełnia ją danymi targetowymi z zestawu danych iris. 
# Zestaw danych iris umożliwia klasyfikację irysów do jednej z trzech klas na podstawie różnych cech, stąd nazwa 'class'.

df['class'] = df['class'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
# Ta linijka kodu przekształca liczby 0, 1, 2 w kolumnie 'class' na odpowiednie nazwy gatunków irysów za pomocą metody map(). 
# Mapowanie to przekształca dane liczbowe na dane kategoryczne.

sns.pairplot(df, hue='class', diag_kind='hist', palette='Set2')
# Funkcja pairplot() z biblioteki seaborn jest używana do tworzenia wykresu relacji par między różnymi cechami w zestawie danych. Hue='class' oznacza,
# że kolor punktów jest określany przez klasę irysów, diag_kind='hist' oznacza, że na przekątnych wykresu znajdują się histogramy cech,
# a palette='Set2' określa schemat kolorów używany do rysowania wykresu.

plt.suptitle("Relacje między cechami Irysów", y=1.02)
# Metoda suptitle() z biblioteki matplotlib jest używana do dodania tytułu do całego wykresu. 
# Argument 'Relacje między cechami Irysów' jest tytułem, a y=1.02 ustawia pozycję tytułu.

plt.show()
# Pokazuje wygenerowany wykres. Bez tej linijki kodu, wykres nie zostanie wyświetlony.
```

# Wnioski, na podstawie zebranych próbek: 
- Iris Setosa wyraźnie odbiega wymiarami od pozostałych dwóch gatunków Irisów.
- Irysy Versicolor i Virginica w przypadku płatków wykazją częśćiowe nakładanie się pomiarów.
- Występuje wyraźna korelacja między petal length / petal width.
- Korelację można też zaobserwować w : sepal length / petal width oraz petal length / sepal length, co jeszcze bardziej uwidacznia poniższa tablica korelacji.

```python
plt.figure(figsize=(14,8))
# Pierwsza linia kodu ustawia rozmiar wykresu na 14x8 cali. Używamy do tego funkcji figure z biblioteki matplotlib.pyplot, której skrótem jest plt. 
# W argumencie figsize podajemy wymiary wykresu.

sns.heatmap(df.drop('class', axis=1).corr(), annot=True, cmap='coolwarm', fmt=".2f")
# Następnie używamy funkcji heatmap z biblioteki seaborn, która jest oznaczona jako sns. Ta funkcja tworzy mapę ciepła, która jest wykorzystana do wizualizacji macierzy korelacji. 
# Tworzymy macierz korelacji dla DataFrame 'df', ale najpierw z niego usuwamy kolumnę 'class' używając metody drop. 
# Argumenty annot i fmt służą do kontrolowania wyglądu etykiet na wykresie - annot=True oznacza, że wartości korelacji będą wyświetlane na wykresie,
# a fmt=".2f" oznacza, że będą one zaokrąglone do dwóch miejsc po przecinku. Argument cmap='coolwarm' określa paletę kolorów używanych na wykresie.

plt.title("Macierz korelacji cech")
# Ta linia kodu ustawia tytuł wykresu na 'Macierz korelacji cech'. Używamy do tego funkcji title z biblioteki matplotlib.pyplot.

plt.show()
# Ostatnia linia kodu wywołuje funkcję show z biblioteki matplotlib.pyplot, która wyświetla utworzony wykres.
```

# 6. Analiza wartości odstających

```python
plt.figure(figsize=(14,8))
# Ta linia kodu tworzy nowy wykres za pomocą funkcji 'figure' z biblioteki matplotlib.pyplot, która jest tutaj skrócona do 'plt'.
# Argument 'figsize' określa rozmiar wykresu, tutaj jest to 14 jednostek szerokości i 8 jednostek wysokości.

sns.boxplot(x='class', y='petal length (cm)', hue='class', data=df, palette='Set3', legend=False)
# Ta linia kodu tworzy pudelkowy wykres (boxplot) za pomocą funkcji 'boxplot' z biblioteki seaborn, która jest tutaj skrócona do 'sns'. 'x' to nazwa kolumny w danych,
# która będzie podstawą wykresu (w tym przypadku, jest to 'class'), 'y' to nazwa kolumny, której dane będą przedstawione na wykresie (tutaj, 'petal length (cm)').
# 'hue='class'' oznacza, że różne klasy będą miały różne kolory na wykresie.
# 'data' to ramka danych (DataFrame), z której będą pobierane dane (tutaj 'df'). 'palette' to paleta kolorów, która ma być użyta do wykresu, tutaj określona jako 'Set3'.
# 'legend=False' oznacza, że na wykresie nie będzie legendy.

plt.title("Porównanie długości płatków między gatunkami")
# Ta linia kodu dodaje tytuł do wykresu - 'Porównanie długości płatków między gatunkami', używając funkcji 'title' z matplotlib.pyplot.

plt.show()
# no i na koniec wywołuję funkcję show pokazującą wykres boxplot
```

```python
plt.figure(figsize=(14,8))

sns.boxplot(x='class', y='petal width (cm)', hue='class', data=df, palette='Set3', legend=False)

plt.title("Porównanie szerokość płatków między gatunkami")

plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.boxplot(x='class', y='sepal length (cm)', hue='class', data=df, palette='Set3', legend=False)

plt.title("Porównanie długość kielichów między gatunkami")

plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.boxplot(x='class', y='sepal width (cm)', hue='class', data=df, palette='Set3', legend=False)

plt.title("Porównanie szerokość kielichów między gatunkami")

plt.show()
```

```python
# Zmiana wykresu z boxplot na scatterplot w celu sprawdzenia który wykres jest czytelniejszy
plt.figure(figsize=(14,8))

sns.scatterplot(x='class', y='sepal length (cm)', hue='class', data=df, palette='Set3', legend=False)

plt.title("Porównanie długość kielichów między gatunkami")

plt.show()
```
